## ¿Cuál es la tasa de negatividad de los medios y cuál es la periodicidad de la negatividad (tasa de negatividad por fecha)?

```python
import elasticsearch
from datetime import datetime
```

```python
country="chile"
from_="2019-09-01"
to_="2019-12-31"
media_outlets=["latercera"]

keyword=""
simple_keyword=False
```

```python
IP = "search.sophia2.org"
PORT = 9200
USER= "elastic"
PASS= "dummy_elastic"
```

```python
#NO MODIFICAR

es = elasticsearch.Elasticsearch(
    IP,
    #port=PORT,
    http_auth=(USER, PASS)
)
```

```python
#NO MODIFICAR
match=""
if (simple_keyword):
    match="match"
else:
    match="match_phrase"

query = { 
    "bool": { 
      #"must": [
      #  {match: { "text":keyword}}

      #],
    "filter": [
        {"range": {
      "date": {
        "gte": from_,
        "lt": to_
      }}},
        { "term":  { "country": country }},
        { "terms":  { "media_outlet": media_outlets }} 
    ]
    }  
}
```

```python
res = es.search(index="news", query=query, size=10000)
print("Son %d noticias encontradas..." % res['hits']['total']['value'])
```

```python
import pandas as pd

data = {'id_news':[],'country':[],'media_outlet':[],'url':[],'title':[],'text':[],'date':[],'search':[]}

df = pd.DataFrame(data)  
  
for hit in res['hits']['hits']:
    id_news = hit['_source']['id_news']
    country = hit['_source']['country']
    media_outlet = hit['_source']['media_outlet']
    url = hit['_source']['url']
    title = hit['_source']['title']
    text = hit['_source']['text']
    date = hit['_source']['date']
    search = keyword
    
    new_row = {'id_news':id_news, 'country':country, 'media_outlet':media_outlet, 'url':url, 'title':title, 'text':text, 'date':date, 'search':search}
    
    df = df.append(new_row, ignore_index=True)

df
```

```python
df['url'].duplicated().any()
```

```python
df = df.drop_duplicates(subset='url', keep='first')
```

```python
nombre_archivo=keyword.replace(' ', '')+"_"+"latercera"+"_"+from_+"_"+to_+".csv"
print("El archivo se llama:")
print(nombre_archivo)
df.to_csv("./"+nombre_archivo)
```

```python
a = pd.read_csv("_latercera_2018-01-01_2018-04-30.csv")
b =  pd.read_csv("_latercera_2018-05-01_2018-09-30.csv")
c =  pd.read_csv("_latercera_2018-10-01_2018-12-31.csv")
d =  pd.read_csv("_latercera_2019-01-01_2019-04-30.csv")
e = pd.read_csv("_latercera_2019-05-01_2019-08-31.csv")

dataset_latercera_2018_2019 = pd.concat([a,b,c,d,e])
```

```python
dataset=dataset_latercera_2018_2019
```

```python
from tqdm import tqdm # Barra de progreso
import transformers
transformers.__version__
```

```python
model_name = "finiteautomata/beto-sentiment-analysis"

from transformers import AutoTokenizer, AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

from transformers import pipeline
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
dataset['title_label'] = ""
dataset['title_score'] = ""

for index, row in tqdm(dataset.iterrows(), desc='Análisis de sentimiento', total=dataset.shape[0]):
    
    # Analizamos su sentimiento en inglés
    sentiment_value = nlp(row['title'])
    
    # Insertamos en dataframe
    dataset.at[index, "title_label"] = sentiment_value[0].get('label')
    dataset.at[index, "title_score"] = sentiment_value[0].get('score')
```

```python
from pandasql import sqldf 
```

```python
## Cantidad de noticias por mes y año

result = sqldf("SELECT date, title_label, count(*) FROM dataset GROUP BY title_label, date ORDER BY date ASC")

result
```
